Summary

The final dataframe is a merger between 2 datasets (‘Graduate Rate’ and ‘Ed Stats Data’) using ‘graduate rate' as the bases of the dataframe and ‘ed stats' for additional features needed. I began by exploring both datasets in which I noticed that Ed Stats only went up to the year 2014, I decided to use it as stat data since it was the most recent year. To merge both datasets, the effective way to join them is by nationality. The two additional features of interest were “adj net enrollment of lower secondary” and “ tertiary educ enrollment(all programs)”.  To merge the datasets a left join was performed. After each merge was completed, I checked if either column had any null values. Since “adj net enrollment of lower secondary” had 66 null values and “tertiary educ enrollment(all programs)” had 19 null values I imputed 0 for values since about less than 2% of the data had null values.
