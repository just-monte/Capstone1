# Imports 

```python
import pandas as pd
import numpy as np
```

# Loading the data 

```python
student_data = pd.read_csv(r'C:\Users\Desktop\USF-DS\educationstats.csv')
student_data.head()
```

# Dataframe 

```python
student_data.info()
```

```python
student_data = student_data.drop(columns=['Country Name'])
```

###### Changing Target column values into numerical 

```python
# Changing Target value "Graduate" to "1"
student_data['Target'] = student_data['Target'].replace(['Graduate'], '1')
```

```python
# Changing Target value "Dropout" to "0"
student_data['Target'] = student_data['Target'].replace(['Dropout'], '0')
```

```python
# Changing Target value "Enrolled" to "2"
student_data['Target'] = student_data['Target'].replace(['Enrolled'], '2')
```

```python
student_data.head()
```

# Saving the dataframe 

```python
student_data.to_csv('student_data.csv')
```

# Summary 

After analyzing the dataframe, I decided to drop the "Country Name" column as there is already a "Nationality" column identifying each participant. Also noticed that the "Target" was the only non-numeric column. I changed the 'Target" values to numeric: "1" for graduate, "2" for enrolled, and "0" for dropout.
